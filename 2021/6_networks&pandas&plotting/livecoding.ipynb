Pandas mapping function to assign a number to each flower variety

```python
color = {'Setosa': 0,
         'Versicolor':1,
         'Virginica':2}

df["color"]= df["variety"].map(color)
```

Loop through all possible combinations of scatter plots and use histogram for diagonal.

```python
fig, axes = plt.subplots(figsize = (16,12),ncols = 4, nrows = 4)
print(axes.shape)

for i, row in enumerate(axes):
    for j, col in enumerate(row):
        if i == j:
            col.hist(df.iloc[:,i])
            col.set_xlabel(df.columns[i])
        else:
            col.scatter(df.iloc[:,i],df.iloc[:,j], c=df["color"])
            col.set_xlabel(df.columns[i])
            col.set_ylabel(df.columns[j])
```
